# Matplotlib
Matplotlib is an amazing visualization library in Python for 2D plots of arrays.
One of the greatest benefits of visualization is that it allows us visual access to huge amounts of data in easily digestible visuals.
Matplotlib consists of several plots like line,bar,scatter,histogrametc.

```python
pip install matplotlib
```

```python
#single graph ploting
import matplotlib.pyplot as plt
x=[5,2,9,4,7]
y=[10,5,8,4,2]
plt.plot(x,y)
plt.title('amplitude vs time')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.show()
```

```python
#Multiline plot
import matplotlib.pyplot as plt
month=['jan','feb','mar','apr','may']
pnb=[515.00,222.70,113.00,612.00,210.00]
sbi=[420.00,730.00,820.55,420.25,910.80]
plt.plot(month,pnb,'g',label='pnb',linewidth=5)
plt.plot(month,sbi,'b',label='sbi',linewidth=6)
plt.title('top banking sector in india by market capitalisation')
plt.xlabel('month')
plt.ylabel('sharemarket')
plt.legend()
plt.savefig("plot1.jpg",dpi=90)
plt.show()
```

Bar Plot

```python
#bar graph
import matplotlib.pyplot as plt
x=[5,2,9,4,7]
y=[10,5,8,4,2]

plt.bar(x,y)
plt.title('amplitude vs time')
plt.xlabel('time')
plt.ylabel('amplitude')

plt.show()
```

pie chart

```python
#pie chart
import matplotlib.pyplot as plt
Car=['Audi','BMW','Tesla','Jaguar']
data=[23,35,29,41]
plt.pie(data, labels=Car)
plt.show()
```

```python
#pie chart with percentage
import matplotlib.pyplot as plt
Car=['Audi','BMW','Tesla','Jaguar']
data=[23,35,29,41]
plt.pie(data, labels=Car, autopct='%1.1f%%')

plt.show()
```

Scatter Chart

```python
#Scatter Chart
import matplotlib.pyplot as plt
x=[5,2,9,4,7]
y=[10,5,8,4,2]
plt.scatter(x, y)
plt.title('amplitude vs time')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.show()
```

subplots()

```python
#subplots()
import matplotlib.pyplot as plt
month=['jan','feb','mar','apr','may']
pnb=[515.00,222.70,113.00,612.00,210.00]
sbi=[420.00,730.00,820.55,420.25,910.80]
plt.plot(month,pnb)
plt.subplot(1, 2, 2)
plt.bar(month, sbi, label="pnb", color='blue')
plt.title("pnb")
plt.subplot(1, 2, 2)
plt.bar(month, sbi, label="SBI", color='green')
plt.title("SBI")
plt.show()
```
